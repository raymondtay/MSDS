**Part C)** Find the period for each state.

Save your answer as a vector called <code>p1.c</code> of length $6$, containing the period of states $1$, $2$, $3$, $4$, $5$, and $6$, respectively.

```r
p1.c = NA
# your code here
# Period of Each State in a Markov Chain
# ---------------------------------------
# Period of state i = gcd{ n >= 1 : P^n(i,i) > 0 }
# i.e. the GCD of all return times with positive probability.
#
# For irreducible chains, all states share the same period.
# For reducible chains, states in the same communication class
# share the same period, but different classes can differ.

gcd <- function(a, b) {
  while (b != 0) {
    temp <- b
    b <- a %% b
    a <- temp
  }
  return(a)
}
 
# --- Compute period of each state ---
# Period(i) = gcd{ n >= 1 : P^n[i,i] > 0 }
# We compute P^1, P^2, ..., P^max_power and collect return times.
# For an n-state chain, the period divides n, so checking up to
# ~2*n^2 powers is more than sufficient for the GCD to stabilise.
 
max_power <- 30
Pn <- diag(n)          # P^0 = I
periods <- rep(0, n)   # will hold the period for each state
 
# Store return times for display
return_times <- vector("list", n)
for (i in 1:n) return_times[[i]] <- integer(0)
 
for (k in 1:max_power) {
  Pn <- Pn %*% P       # P^k
  for (i in 1:n) {
    if (Pn[i, i] > 1e-15) {
      return_times[[i]] <- c(return_times[[i]], k)
    }
  }
}
 
# Compute GCD of return times for each state
for (i in 1:n) {
  rt <- return_times[[i]]
  if (length(rt) == 0) {
    periods[i] <- NA     # state never returns (shouldn't happen for recurrent)
  } else if (length(rt) == 1) {
    periods[i] <- rt[1]
  } else {
    g <- rt[1]
    for (j in 2:length(rt)) {
      g <- gcd(g, rt[j])
      if (g == 1) break  # GCD=1 is minimal, no need to continue
    }
    periods[i] <- g
  }
}
 
# --- Display results ---
cat("Return times P^n[i,i] > 0 for n = 1..30:\n")
cat(strrep("-", 55), "\n")
for (i in 1:n) {
  rt <- return_times[[i]]
  # Show first few return times to keep output readable
  display <- if (length(rt) > 10) {
    paste0(paste(rt[1:10], collapse = ", "), ", ...")
  } else {
    paste(rt, collapse = ", ")
  }
  cat(sprintf("  State %d: {%s}  =>  period = %d\n", i, display, periods[i]))
}
 
# --- Verify: all states in same class must share the same period ---
# (reuse classes from previous script)
A <- (P > 0) * 1
R <- diag(n)
power <- A
for (k in 1:(n - 1)) {
  R <- R + power
  power <- power %*% A
}
R <- (R > 0) * 1
C <- R * t(R)
 
visited <- rep(FALSE, n)
classes <- list()
for (i in 1:n) {
  if (!visited[i]) {
    members <- which(C[i, ] == 1)
    classes <- c(classes, list(members))
    visited[members] <- TRUE
  }
}
 
cat("\n\nVerification — period by communicating class:\n")
cat(strrep("-", 55), "\n")
for (k in seq_along(classes)) {
  members <- classes[[k]]
  p <- unique(periods[members])
  cat(sprintf("  Class %d: {%s}  =>  period = %d\n",
              k, paste(members, collapse = ", "), p))
}
```
